# **Exercícios utilizando Listas Python**

## 1. Soma dos Elementos da Lista
Dada uma lista de números, calcule a soma de todos os elementos.

```python
# Seu código aqui...
lista = [1, 2, 3, 4, 5]
soma_de_numeros = 0
for listas in lista:
  soma_de_numeros += listas
print(soma_de_numeros)
```

## 2. Produto dos elementos da lista
Dada uma lista de números, calcule o produto de todos os elementos.

```python
# Seu código aqui...
def produto_lista(lista):
    produto = 1
    for numero in lista:
        produto *= numero
    return produto

lista2 = [1, 2, 3, 4, 5]
resultado = produto_lista(lista2)
print("O produto dos elementos da lista é:", resultado)
```

## 5. Contar Ocorrências de um Elemento
Dada uma lista de números e um número alvo, conte quantas vezes o alvo aparece na lista.

```python
# Seu código aqui...
def contar_ocorrencias(lista5, alvo):
    return lista5.count(alvo)

lista5 = [1, 2, 3, 4, 3, 2, 1]
alvo = 2
print(f"O número {alvo} aparece {contar_ocorrencias(lista5, alvo)} vezes na lista.")
```

## 6. Remover duplicatas
Dada uma lista de números, crie uma nova lista sem elementos duplicados.

```python
# Seu código aqui...
lista61 = [1, 2, 2, 3, 4, 4, 5]
lista62 = []
for sduplicado in lista61:
  if sduplicado not in lista62:
    lista62.append(sduplicado)
print(lista62)
```

## 8. Intercalar Duas Listas
Dadas duas listas de números, crie uma nova lista intercalando os elementos das duas listas.

```python
# Seu código aqui...
lista81 = [1, 3, 5]
lista82 = [2, 4, 6]
lista83 = []

for item1, item2 in zip(lista81, lista82):
  lista83.append(item1)
  lista83.append(item2)

print(lista83)
```

## 9. Somar Elementos Correspondentes de Duas Listas
Dadas duas listas de números de mesmo tamanho, crie uma nova lista com a soma dos elementos correspondentes.

```python
# Seu código aqui...
def somar_elementos_correspondentes(lista91, lista92):
    return [x + y for x, y in zip(lista91, lista92)]

lista91 = [1, 2, 3]
lista92 = [4, 5, 6]
lista93 = somar_elementos_correspondentes(lista91, lista92)
print(f"A nova lista com a soma dos elementos correspondentes é: {lista93}")
```

## 10. Mover Zeros para o Fim
Dada uma lista de números, mova todos os zeros para o final da lista.

```python
# Seu código aqui...
lista10 = [0, 1, 0, 3, 12]
lista10 = [num for num in lista10 if num != 0 ] + [0]
print(lista10)
```
